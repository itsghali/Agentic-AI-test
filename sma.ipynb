```python
import tiktoken
from dotenv.ipython import load_dotenv
import os
```

```python
tokennizer=tiktoken.encoding_for_model("gpt-4o")
print(tokennizer.name)
```

```python
prompt="What is the capital of France?"
tokens=tokennizer.encode(prompt)
print(tokens)
```

```python
for token in tokens:
    t=tokennizer.decode_single_token_bytes(token).decode("utf-8")
    print(t)
```

```python
def count_tokens(prompt:str, model:str="o200k_base"):
    tokennizer=tiktoken.get_encoding(model)
    return len(tokennizer.encode(prompt))
```

```python
print(count_tokens(prompt="dkdkd kdkd"))
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-4o", temperature=0)
```

```python
resp=llm.invoke(input=[
    {"role":"system", "content":"You are a helpful assistant."},
    {"role":"user", "content":"What is the definition of artificial intelligence with tables and examples? "}    
])
```

```python
from IPython.display import Markdown
print(display(Markdown(resp.content)))
```
